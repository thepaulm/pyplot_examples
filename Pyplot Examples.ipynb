```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
domain = np.linspace(-np.pi, np.pi, 200)
```

```python
s = np.sin(domain)
```

```python
c = np.cos(domain)
```

```python
fig = plt.figure(1, figsize=(10, 4))
```

```python
ax1 = fig.add_subplot(2, 1, 1)
```

```python
ax1.plot(s)
```

```python
ax2 = fig.add_subplot(2, 1, 2)
```

```python
ax2.plot(c)
```

```python
def plot_them_all(*args):
    styles = ['r-', 'bo', 'g--']
    count = len(args)
    w = 20
    h = 4 * count
    fig = plt.figure(1, figsize=(w, h))
    for i, a in enumerate(args):
        ax = fig.add_subplot(count, 1, i+1)
        style = styles[i % len(styles)]
        ax.plot(a, style)
```

```python
plot_them_all(s, c, s, c)
```
