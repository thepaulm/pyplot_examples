```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
domain = np.linspace(-np.pi, np.pi, 200)
```

```python
s = np.sin(domain)
```

```python
c = np.cos(domain)
```

```python
def plot_them_all(*args):
    styles = ['r-', 'bo', 'g--']
    count = len(args)
    w = 20
    h = 4 * count
    fig = plt.figure(1, figsize=(w, h))
    for i, a in enumerate(args):
        ax = fig.add_subplot(count, 1, i+1)
        style = styles[i % len(styles)]
        ax.plot(a, style)
```

```python
plot_them_all(s, c, s, c)
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
def pta(*args, color_rotate=True):
    '''pta = plot them all'''
    count = len(args)
    w = 20
    h = 4 * count
    fig = plt.figure(1, figsize=(w, h))
    for i, a in enumerate(args):
        ax = fig.add_subplot(count, 1, i+1)
        if color_rotate:
            for _ in range(i):
                ax.plot([],[])
        ax.plot(a)
```

```python
pta(c, s, s, c)
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
pta(c, s, s, c, color_rotate=False)
```

```python
def ptt(*args):
    '''ptt = plot them together'''
    count = len(args)
    w = 20
    h = 4
    fig = plt.figure(1, figsize=(w, h))
    ax = fig.add_subplot(1, 1, 1)
    for a in args:
        ax.plot(a)
```

```python
ptt(c, s)
```
